```python
fignum =1

for title, regularizer in (('Soft-Margin', 10), ('Hard-Margin', 100)):
    clf = svm.SVC(kernel='linear', C=regularizer,random_state=2019)
    clf.fit(predictors, target)

    print("number of support vectors for %s: %d for class 0 and %d for class 1" %(title,clf.n_support_[0],clf.n_support_[1]))
    print("number of support vectors for %s: %d for class 0 and %d for class 1" %(title,clf.n_support_[0],clf.n_support_[1]))

 
    # get the separating hyperplane
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(-1, 1)
    yy = a * xx - (clf.intercept_[0]) / w[1]

    # plot the parallels to the separating hyperplane that pass through the
    # support vectors (margin away from hyperplane in direction
    # perpendicular to hyperplane). This is sqrt(1+a^2) away vertically in
    # 2-d.
    margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
    yy_down = yy - np.sqrt(1 + a ** 2) * margin
    yy_up = yy + np.sqrt(1 + a ** 2) * margin

    # plot the line, the points, and the nearest vectors to the plane
    plt.figure(fignum, figsize=(8, 6))
    plt.clf()
    plt.plot(xx, yy, 'k-')
    plt.plot(xx, yy_down, 'k--')
    plt.plot(xx, yy_up, 'k--')

    plt.xlim(-1, 1)
    plt.ylim(-2, 1.5)
    
    plt.title(title)
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                    facecolors='y', zorder=10, edgecolors='y')
    plt.scatter(predictors[:, 0], predictors[:, 1], c=target, zorder=10, cmap=plt.cm.Paired,
                    edgecolors='k')

    fignum+=1
#plt.show()
```

As we see that we have more support vectors for soft margin svm (4+2) 6 and only (1+2) 3 support vector for the hard margin. this is due to the fact that the hard margin doesn't allow the outlier point to be misclassified so the hyperplane goes far from the points detected by soft margin and looks for the outlier.

We see from each dataset plot that :

Dataset one is presenting a linearly separable problem (as we've seen from Task4)
Dataset Two seems to need a polynomial kernel but we'll need to find it's degree
Dataset three is clearly suggesting a radial kernel
We could try different kernels for each dataset, but we will use our knowledge from data visualisation to gain processing time. So we will use :

Linear for Dataset1
RBF + Linear for Dataset 2
Poly+Linear for Dataset3

```python
def Tunning_SVC(ds,kernel):


    X=ds.iloc[:,:2].values
    y=ds.iloc[:,2]


    # Split the dataset in two equal parts
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=2019)

    # Set the parameters by cross-validation
    if kernel=='linear':
        tuned_parameters = [{'kernel': [kernel], 'C': [1, 10, 100, 1000]}]
        
    elif kernel=='rbf':
        tuned_parameters = [{'kernel': [kernel,'linear'], 'gamma': [1e-3, 1e-4],
                         'C': [1, 10, 100, 1000]}]
        
    elif kernel=='poly':
        tuned_parameters = [{'kernel': [kernel,'linear'], 'gamma': [1e-3, 1e-4],
                         'C': [1, 10, 100, 1000],'degree':[2,3,4,5,6]}]

    scores = ['f1'] # F1 score was used due to imbalanced classes in target features for datasets

    for score in scores:
        print("# Tuning hyper-parameters for %s" % score)
        print()

        clf = GridSearchCV(svm.SVC(), tuned_parameters, cv=5, scoring='%s_macro' % score)
        clf.fit(X_train, y_train)

        print("Best parameters set found on development set:")
        print()
        print(clf.best_params_)
        print()
        print("Grid scores on development set:")
        print()
        means = clf.cv_results_['mean_test_score']
        stds = clf.cv_results_['std_test_score']
        for mean, std, params in zip(means, stds, clf.cv_results_['params']):
            print("%0.3f (+/-%0.03f) for %r"
                  % (mean, std * 2, params))
        print()

        print("Detailed classification report:")
        print()
        print("The model is trained on the full development set.")
        print("The scores are computed on the full evaluation set.")
        print()
        y_true, y_pred = y_test, clf.predict(X_test)
        print(classification_report(y_true, y_pred))
        print()
```

# Dataset 1

```python
Tunning_SVC(ds=dataset1,kernel='linear')
```

# Dataset 2

```python
Tunning_SVC(ds=dataset2,kernel='poly')
```

# Dataset 3

```python
Tunning_SVC(ds=dataset3,kernel='rbf')
```

For Dataset 1 best model based on F1 score is obtained with {'C': 1, 'kernel': 'linear'}. Dataset 2 was processed with both Linear and poly kernels, and suprisingly linear kernel fits a better model than polynomial one which is perhaps related to overfitting, so we conclude that it was a good idea to try linear kernel even if the visualisation suggested a polynomial one. Regarding Dataset 3, as expected, RBF is the best kernel giving top model score with the following hyperparameters : {'C': 1, 'gamma': 0.001, 'kernel': 'rbf'}
