```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import math

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings('ignore')
```

```python
def load_data(filename):
    X_load = np.genfromtxt('./'+filename, delimiter=',', skip_header=1, usecols = (1, 2))
    y_load = np.genfromtxt('./'+filename, delimiter=',', skip_header=1, usecols = (3))
    return X_load, y_load
```

### Task 4
Perform classification on dataset_1 using an SVM with linear kernel and experiment with the C-parameter to find the widest margin solution with a hard margin and a soft margin. 

```python
def plot_decision_boundary(classifier, XXX_train, yyy_train, title=None):
    plt.scatter(XXX_train[:, 0], XXX_train[:, 1], c=yyy_train, s=30, cmap=plt.cm.Paired)
    plt.title(title)
    # plot the decision function
    ax = plt.gca()
    ax.set_aspect(aspect=1)
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = classifier.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    # plot support vectors
    ax.scatter(classifier.support_vectors_[:, 0], classifier.support_vectors_[:, 1], s=100,  
               linewidth=1, facecolors='b', edgecolors='y')

def print_metrics(classifier, C, score):
    header = "="*10 + " SVM with C=" + str(C) + " " + "="*10
    print(header)
    print("Nb support vectors: ", len(clf.support_vectors_))
    print("Nb support vectors per class:", clf.n_support_ )
    print("Score :", score)
    return len(clf.support_vectors_)
```

```python
C_vals = [0.01, 0.1, 1, 10, 50, 100, 1000]
tuned_params = []
# Using a SVM classifier
SVM_clf = svm.SVC()
# Add a figure for plotting the decision boundarys in subplots
f = plt.figure(figsize=(20,20))

nb_svms = []

for idx, value in enumerate(C_vals):
    tune_param = {'kernel': ['linear'], 'C': [value]}
    tuned_params.append(tune_param)
    classifier = GridSearchCV(SVM_clf, tune_param, scoring='accuracy')
    classifier.fit(X_train, y_train)
    ax = f.add_subplot(1,len(C_vals),idx+1)
    nb_support_vectors = len(classifier.best_estimator_.support_vectors_)
    nb_svms.append(nb_support_vectors)
    plot_decision_boundary(classifier.best_estimator_, X_train, y_train, 'C value={}, #SV={}'.format(value,nb_support_vectors))
    y_pred = classifier.best_estimator_.predict(X_test)
    score = classifier.best_estimator_.score(X_test, y_test)
    print("SVM classifier parameters:{} with validation score={}, test score={}, support vector count={}, vectors by class={}".format(classifier.best_params_,
                                                                          classifier.best_score_, score,nb_support_vectors, classifier.best_estimator_.n_support_ ))                      
```

```python
print(C_vals, nb_svms)
plt.plot(C_vals, nb_svms)
plt.ylabel('Nb Support Vectors')
plt.xlabel('C parameter')
plt.title('Number of Support Vectors wrt C')

ax = plt.gca()
ax.semilogx(value, np.exp(-value))
ax.semilogy(np.amax(nb_svms), np.exp(-np.amax(nb_svms)))

ax.grid()
plt.show()
```

The score is consistently at 1 as the data is linearly separable

The isolated point which is the outlier in the dataset is on the bounday for lower values of C and within the margin for high values of C

The number of support vectors identified is inversely proportional to the value of C. This means that smaller the value of C, the hyperplane for classification would consist of many Support vectors and the margin is also wider. This is a case of soft margin. Hence lower the value of C - higher is the risk of underfitting

The large value of C (Hard Margin) only resulted in a small number of support vectors closer to the line of separation. This results in a very narrow margin and is also considered as the hard margin. This is also the case for overfitting.

The medium values of C i.e. nearabouts 10 has a relatively small number of support vectors and the margins are robust as they are parallel to distribution borders.

The plot "Number of Support Vectors wrt C" shows that the number of support vectors decrease as the value of C increases. The lowest number of support vectors is around 3 for the value of C in [50,100]

**Conclusion**
The plots above signify that the median value of C gives us a least number of support vectors and better generalization and hence a lower VC dimension which is desirable.

```python
def evaluate_hyper_parameters(XX_train,yy_train, XX_test, yy_test):
    # using gridsearchCV to choose best parameters
    # Set the parameters by cross-validation
    tuned_parameters = [{'kernel': ['linear'], 'C': [0.1, 1, 10, 20]},
                        {'kernel': ['rbf', 'poly'],
                         'gamma': [0.1, 1, 'auto'],
                         'C': [0.1, 1, 10, 20],
                         'degree': [1, 2]}]
    
    classifier = GridSearchCV(SVM_clf, tuned_parameters, cv=10, scoring='accuracy', n_jobs=-1)
    classifier.fit(XX_train, yy_train)
    plot_decision_boundary(classifier.best_estimator_, XX_train, yy_train, 
                           'Best Hyper-params :' + str(classifier.best_params_))

    # Make predictions on test data
    yy_pred = classifier.best_estimator_.predict(XX_test)
    
    # Test accuracy score on classiffiaction
    test_score = classifier.best_estimator_.score(XX_test, yy_test)
    
    # Calculate the Expected Error based on the number of Support Vectors
    nb_sv = len(classifier.best_estimator_.support_vectors_)
    nb_ts = len(X_train)
    Eout = nb_sv/(nb_ts -1)
    
    print("**"*40)
    print("Number of Support Vectors  :", nb_sv)
    print("Number of Training Samples :", nb_ts)
    print("Expected E_out :", Eout)
    print("Best SVM parameters on cross-validation:", classifier.best_params_)
    print("Best CV score :", classifier.best_score_)
    print("Test data score", test_score)
    print("expected[Eout] : ", Eout)
    print("**"*40)    
```

```python
# Determination of best params for dataset1
# split data
evaluate_hyper_parameters(X_train,y_train, X_test, y_test)
```

### Observations
#### Dataset 1
Gridsearch selected the `linear` kernel for dataset_1 since the dataset is linearly separable as it is visible from the plot above.

The most important parameter for the `linear` kernel is `C` 
  * High `C` leading to narrow margins
  * Low `C` leading to large margins at the expense of misclassifications.
  

#### Dataset 2
Gridsearch selected the `rbf` kernel for dataset_2.

Two of the most important parameters for `rbf` are `C` and `gamma`:
  * `gamma` represents how far the influence of a single training example reaches:   
  * `C` behaves as a regularization parameter like for `linear` kernel with:
    * High `C` leading to narrow margins
    * Low `C` leading to large margins
    
Reference: https://scikit-learn.org/stable/auto_examples/svm/plot_rbf_parameters.html

#### Dataset 3
Gridsearch selected the `rbf` kernel as the winner for dataset_3.

We notice that `poly` was not selected despite the data set being suited to 2nd degree transformation as seen in a previous Etivity. 

This leads to conclude that, `rbf` is a good kernel for non-linearly separable datasets
